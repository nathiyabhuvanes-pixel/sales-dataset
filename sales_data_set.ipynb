```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(file_path, encoding='latin1')

# Convert Sale_Date to datetime
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])

# 1. Total Sales

total_sales = df["SALES"].sum()
print("Total Sales:", total_sales)


# 2. Sales by Region

sales_by_region = df.groupby("COUNTRY")["SALES"].sum()
print("\nSales by Region:")
print(sales_by_region)

# 3. Sales by Product Category

sales_by_category = df.groupby("PRODUCTLINE")["SALES"].sum()
print("\nSales by Product Category:")
print(sales_by_category)

# 4. Monthly Sales Trend

df["Month"] = df["ORDERDATE"].dt.to_period("M")
monthly_sales = df.groupby("Month")["SALES"].sum()

# Plot monthly sales
plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.show()

### ð¹ Top Sales Representative

# top_rep = df.groupby("Sales_Rep")["SALES"].sum().sort_values(ascending=False) # 'Sales_Rep' column not found
# print("\nTop Sales Representatives:")
# print(top_rep)


### ð¹ Total Profit

# df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"] # 'Unit_Price', 'Unit_Cost', 'Quantity_Sold' columns not found
# total_profit = df["Profit"].sum()
# print("\nTotal Profit:", total_profit)
```
